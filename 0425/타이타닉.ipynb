```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

#### 1. 데이터 전처리

```python
df = df[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked', 'deck']]
```

- 결측치 처리

```python
df.isna().sum()
```

```python
#age na - 평균으로 대체
df.age.fillna(df.age.mean(), inplace=True)
df.age.isnull().sum()
```

```python
# embarked na - 최진값으로 대체
df.embarked.value_counts()
```

```python
df.embarked.fillna('S', inplace=True)
df.embarked.isnull().sum()
```

```python
# deck na - 너무 많은 데이터가 na 이므로 열을 삭제하는 것이 맞음
df.drop(columns=['deck'], inplace=True)
```

```python
df.isna().sum()
```

- 카테고리 값(sex, embarked)를 숫자로 변환

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.sex = le.fit_transform(df.sex)
df.embarked = le.fit_transform(df.embarked)
df.head()
```

#### 2. Train/Test Dataset으로 분리

```python
# X와 y를 np 배열로 변환
X = df.iloc[:, 1:].values #survived 제외 모든 값
y = df.survived.values
X.shape, y.shape
```

```python
# y값의 분포
# 방법 1. df.survived.value_counts()
np.unique(y, return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, stratify=y, test_size=0.2, random_state=2022)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
np.unique(y_train, return_counts=True)
```

#### 3. RandomForest 모델로 학습시키기

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=2022)
rfc.get_params()
```

```python
rfc.fit(X_train, y_train)
```

#### 4. 모델 예측 및 평가

```python
rfc.score(X_test, y_test)
```

#### 5. 3, 4 대신에 GridSearchCV 수행

```python
params={
    'max_depth':[2,4,6,8],
    'min_samples_split':[2,4,6]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_rf=GridSearchCV(rfc, params, scoring='accuracy', cv=5)
grid_rf.fit(X_train, y_train)
```

```python
grid_rf.best_params_
```

```python
params={
    'max_depth':[3,4,5],
    'min_samples_split':[3,4,5]
}
grid_rf=GridSearchCV(rfc, params, scoring='accuracy', cv=5)
%time grid_rf.fit(X_train, y_train)
```

```python
grid_rf.best_params_
```

```python
best_rf = grid_rf.best_estimator_
best_rf.score(X_test, y_test)
```

#### 6. 테스트 데이터에 적용

#### 7. 엉터리 분류기 : 여성이면 생존으로 예측하기

```python
# 여성의 생존률 -> 0과 1로 나뉘어져 있으므로 mean()을 쓰면 가능
# sex 0은 여성, 1은 남성
# 여성의 생존률 -> 0.742038
df.groupby('sex')[['survived']].mean()
```

```python
df.groupby(['sex', 'pclass'])[['survived']].mean()
```

```python
# 자체 분류기 만들어보기 -> 상속
from sklearn.base import BaseEstimator

# fit, predict method만 재정의(Overriding)
class MyClassifier(BaseEstimator):
  def fit(self, X, y):
    pass # 엉터리이기 때문에 그냥 pass
  
  def predict(self, X):
    pred = np.zeros(X.shape[0], int) #X의 행의 갯수만큼 0로 초기화한 배열을 생성
    for i in range(X.shape[0]):
      if X[i, 1]==0:            # 여성이면
        pred[i]=1               # 생존
    return pred
```

```python
my_clf=MyClassifier()
my_clf.fit(X_train, y_train)
pred_my = my_clf.predict(X_test)
```

```python
X_test[8,1], pred_my[8]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred_my)
```

```python
pred_rf = best_rf.predict(X_test)
sdf = pd.DataFrame({'y_test':y_test, 'RF':pred_rf, 'My':pred_my})
sdf.head()
```

```python
# 결론 : 모델의 성능을 평가할 때 무조건적으로 정확도를 사용하는 것은 지양해야 한다
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred_rf) # [TN, FP], [FN,TP]
```

```python
confusion_matrix(y_test, pred_my)
```

```python
# 정밀도 (precision score : TP / (FP+TP))
from sklearn.metrics import precision_score, recall_score
precision_score(y_test, pred_rf), precision_score(y_test, pred_my)
```

```python
# 재현율 (recall score : TP / (FN+TP))
recall_score(y_test, pred_rf), recall_score(y_test, pred_my)
```

```python
# F1 스코어 : 정밀도와 재현율의 조화 평균
from sklearn.metrics import f1_score
f1_score(y_test, pred_rf), f1_score(y_test, pred_my)
```

```python
# AUC 스코어
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, pred_rf), roc_auc_score(y_test, pred_my)
```
