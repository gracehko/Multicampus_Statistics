```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE']=boston.target
df.head()
```

- Feature 별 산점도와 선형회귀선

```python
sns.regplot(x='CRIM', y='PRICE', data=df);
```

```python
_, axs = plt.subplots(3, 4, figsize=(16,12))
for i, feature in enumerate(df.columns[1:-1]):
  row, col = i//4, i%4
  sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

- Feature 별 결정계수(R squared)

```python
from sklearn.linear_model import LinearRegression
```

```python
for feature in df.columns[:-1]:
  lr = LinearRegression()
  X = df[feature].values.reshape(-1,1)
  lr.fit(X, boston.target)
  score = lr.score(X, boston.target)
  print(f'{feature}:\t{score:.4f}')
```

- 다중 선형 회귀

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size=0.1, random_state=2022
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# R squared

lr.score(X_train, y_train)
```

```python
# 계수(weight)
lr.coef_
```

```python
# 절편(bias)
lr.intercept_
```

- 회귀식
<pre>
y = -1.18*CRIM + 4.72*ZN + 4.93*INDUS + ... – 5.27 *LSTAT

- 테스트 데이터에 적용

```python
X_test[0], y_test[0]
```

```python
# weight와 bias를 이용하여 직접 계산한 값
np.dot(lr.coef_, X_test[0]) + lr.intercept_
```

```python
# predict method - 위의 결과값과 동일

# lr.predict(X_test[:1]) 도 가능
lr.predict(X_test[0].reshape(1,-1))
```

```python
# 10개의 테스트에 대해 적용
for i in range(10):
  pred1 = np.dot(lr.coef_, X_test[i]) + lr.intercept_
  pred2 = lr.predict(X_test[i].reshape(1,-1))
  print(f'실제값:{y_test[i]},\t직접 계산한 예측값1:{pred1:.4f},\tLR 예측값2:{pred2[0]:.4f}')
```

#### 회귀는 반드시 선형 회귀를 써야 하는가?
- 그 외 결정트리, SVM, RF, XGBoost 등등



```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# 선형회귀
pred_lr = lr.predict(X_test)
r2_lr = r2_score(y_test, pred_lr)
mse_lr = mean_squared_error(y_test, pred_lr)
```

```python
# 결정트리(regressor는 예측가능)
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=2022)
dtr.fit(X_train, y_train)
```

```python
pred_dt = dtr.predict(X_test)
r2_dt = r2_score(y_test, pred_dt)
mse_dt = mean_squared_error(y_test, pred_dt)
```

```python
# SVM
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
```

```python
pred_sv = svr.predict(X_test)
r2_sv = r2_score(y_test, pred_sv)
mse_sv = mean_squared_error(y_test, pred_sv)
```

```python
# RF(랜덤 포레스트)
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=2022)
rf.fit(X_train, y_train)
```

```python
pred_rf = rf.predict(X_test)
r2_rf = r2_score(y_test, pred_rf)
mse_rf = mean_squared_error(y_test, pred_rf)
```

```python
# XGBoost
from xgboost import XGBRegressor
xgb = XGBRegressor(random_state=2022)
xgb.fit(X_train, y_train)
```

```python
pred_xgb = xgb.predict(X_test)
r2_xgb = r2_score(y_test, pred_xgb)
mse_xgb = mean_squared_error(y_test, pred_xgb)
```

```python
# 비교하기 : r2는 클수록, mse는 작을수록 좋다
print('LR\tDT\tSVM\tRF\tXGB')
print(f'{r2_lr:.4f}\t{r2_dt:.4f}\t{r2_sv:.4f}\t{r2_rf:.4f}\t{r2_xgb:.4f}\t')
print(f'{mse_lr:.4f}\t{mse_dt:.4f}\t{mse_sv:.4f}\t{mse_rf:.4f}\t{mse_xgb:.4f}\t')
```
