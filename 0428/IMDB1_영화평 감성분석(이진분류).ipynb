```python
# 파일 : https://www.kaggle.com/competitions/word2vec-nlp-tutorial/data?select=labeledTrainData.tsv.zip
df = pd.read_csv('/content/labeledTrainData.tsv', sep='\t')
df.head()
```

```python
df.info()
```

```python
df.review[0][:1000] #줄바꿈이 <br>
```

- 텍스트 전처리

```python
# <br> 태그를 공백으로 변환
df.review = df.review.str.replace('<br />','')
```

```python
df.review[0][:1000]
```

```python
# 구둣점, 숫자 제거 - 영문자가 아닌 글자는 공백으로 변환 : 정규표현식
# df의 str method는 정규표현식을 지원 / 일반 str는 지원X

import re
df.review = df.review.str.replace('[^A-Za-z]',' ').str.strip()
df.review[0][:1000]
```

- Train/Test dataset 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.review, df.sentiment, stratify=df.sentiment, random_state=2022
)
y_train.value_counts()
```

- CountVectorizer로 변환

```python
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer(stop_words='english')
```

```python
cvect.fit(X_train)
X_train_cv = cvect.transform(X_train)
X_train_cv.shape
```

```python
# test dataset은 train dataset을 변환한 Vectorizer로 변환해야 한다
X_test_cv = cvect.transform(X_test)
X_test_cv.shape
```

- 분류기 : Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
%time nb.fit(X_train_cv, y_train)
```

```python
nb.score(X_test_cv, y_test)
```

- ngram range = (1,2)

```python
cvect2 = CountVectorizer(ngram_range=(1,2), stop_words='english')
cvect2.fit(X_train)
%time X_train_cv2 = cvect2.transform(X_train)
print(X_train_cv2.shape)
```

```python
X_test_cv2 = cvect2.transform(X_test)
X_test_cv2.shape
```

```python
nb2=MultinomialNB()
%time nb2.fit(X_train_cv2, y_train)
```

- 모델 저장하고 불러오기

```python
import joblib
```

```python
joblib.dump(cvect2, 'imdb_cvect_ngram12.pkl')
joblib.dump(nb2, 'imdb_nb2.pkl')
```
